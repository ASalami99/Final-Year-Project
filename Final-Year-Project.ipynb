# Image (Video) Enhancement Techniques

### Histogram Equalization

```python
import cv2
import os
from tqdm import tqdm

input_dir = "/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100"
output_dir_histeq = "/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100_histeq"

def enhance_all_hist_eq(input_dir, output_dir):
    os.makedirs(output_dir, exist_ok=True)
    video_files = [f for f in os.listdir(input_dir) if f.endswith('.mp4')]

    for video_file in tqdm(video_files, desc="Applying Histogram Equalization"):
        input_path = os.path.join(input_dir, video_file)
        output_path = os.path.join(output_dir, video_file)

        cap = cv2.VideoCapture(input_path)
        if not cap.isOpened():
            print(f"❌ Failed to open: {input_path}")
            continue

        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        while True:
            ret, frame = cap.read()
            if not ret:
                break
            ycrcb = cv2.cvtColor(frame, cv2.COLOR_BGR2YCrCb)
            ycrcb[:, :, 0] = cv2.equalizeHist(ycrcb[:, :, 0])
            enhanced = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)
            out.write(enhanced)

        cap.release()
        out.release()
        print(f"✅ Processed: {video_file}")

enhance_all_hist_eq(input_dir, output_dir_histeq)
```

### CLAHE

```python
import cv2
import os
from tqdm import tqdm

input_dir = "/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100"
output_dir_clahe = "/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100_clahe"

def enhance_all_clahe(input_dir, output_dir, clip_limit=2.0, tile_grid_size=(8, 8)):
    os.makedirs(output_dir, exist_ok=True)
    video_files = [f for f in os.listdir(input_dir) if f.endswith('.mp4')]

    for video_file in tqdm(video_files, desc="Applying CLAHE Enhancement"):
        input_path = os.path.join(input_dir, video_file)
        output_path = os.path.join(output_dir, video_file)

        cap = cv2.VideoCapture(input_path)
        if not cap.isOpened():
            print(f"❌ Failed to open: {input_path}")
            continue

        fourcc = cv2.VideoWriter_fourcc(*'mp4v')
        fps = cap.get(cv2.CAP_PROP_FPS)
        width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
        height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
        out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

        clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid_size)

        while True:
            ret, frame = cap.read()
            if not ret:
                break

            # Convert to YCrCb
            ycrcb = cv2.cvtColor(frame, cv2.COLOR_BGR2YCrCb)
            y_channel = ycrcb[:, :, 0]

            # Apply CLAHE to Y channel
            y_channel_clahe = clahe.apply(y_channel)

            # Merge and convert back
            ycrcb[:, :, 0] = y_channel_clahe
            enhanced = cv2.cvtColor(ycrcb, cv2.COLOR_YCrCb2BGR)

            out.write(enhanced)

        cap.release()
        out.release()
        print(f"✅ Processed: {video_file}")

# Run the enhancement
enhance_all_clahe(input_dir, output_dir_clahe)
```

```python
import os
import cv2
import torch
import numpy as np
from PIL import Image
from tqdm import tqdm
from realesrgan import RealESRGANer
from basicsr.archs.rrdbnet_arch import RRDBNet

# Input and output directories
input_dir = '/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100'
output_dir = '/content/drive/MyDrive/Final-Year-Project/archive/data/WLASL100_realesrgan'
os.makedirs(output_dir, exist_ok=True)

# Load RealESRGAN model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = RRDBNet(num_in_ch=3, num_out_ch=3, num_feat=64, num_block=23, num_grow_ch=32, scale=4)
upscaler = RealESRGANer(
    scale=4,
    model_path='/content/drive/MyDrive/Project_with_DS/RealESRGAN_x4plus.pth',
    model=model,
    tile=0,
    tile_pad=10,
    pre_pad=0,
    half=torch.cuda.is_available(),
    device=device
)

# Process each video
video_files = [f for f in os.listdir(input_dir) if f.endswith('.mp4')]
for video_file in tqdm(video_files, desc="Enhancing videos with RealESRGAN"):
    input_path = os.path.join(input_dir, video_file)
    output_path = os.path.join(output_dir, video_file)

    # Skip already processed videos
    if os.path.exists(output_path):
        #print(f"⏭️ Skipping (already processed): {video_file}")
        continue

    cap = cv2.VideoCapture(input_path)
    if not cap.isOpened():
        print(f"❌ Failed to open: {input_path}")
        continue

    fps = cap.get(cv2.CAP_PROP_FPS)
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH)) * 4  # RealESRGAN scale
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT)) * 4
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        try:
            img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            sr_img, _ = upscaler.enhance(np.array(img))
            sr_img_bgr = cv2.cvtColor(sr_img, cv2.COLOR_RGB2BGR)
            out.write(sr_img_bgr)
        except Exception as e:
            print(f"❌ Error processing frame from {video_file}: {e}")

    cap.release()
    out.release()
```
